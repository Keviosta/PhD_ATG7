# I] Dataframe settings

```python
%%time
#Open the big expression data with all the transcripts
path = output_file_directory + "/dataframes_for_input/all_transcripts_zscore.tsv"
df = pandas.read_csv(path, sep = "\t", index_col='sample')

print(df.shape)
df.head()
```

# II] Clusters  - KMeans 

```python
%%time
# 9 is the best number of cluster
kmeans_model = KMeans(n_clusters = 185, random_state=1, verbose=False).fit(df)
labels = kmeans_model.labels_
# Find what cluster for each gene
pred_cluster = kmeans_model.predict(df)
print(pred_cluster)
#add the prediction in a column
df['xcluster'] = pred_cluster
df.head()
```

## B) Heatmap with clusters

```python
%%time
#sort the df with the cluster numbers.
df = df.sort_values('xcluster')

#save index of 1st row of each clusters for horizontal line
nb_cluster = 185
horiz_line_pos = []
print('First row for each clusters:')
for i in range(1,nb_cluster):
    
    first_index = df.reset_index().loc[df.reset_index()['xcluster'].isin([i])].index[0]
    horiz_line_pos.append(first_index)

print(horiz_line_pos)

# calculate the number of transcripts per cluster
fst_index_clust = horiz_line_pos
length = len(fst_index_clust)
print(f'\nNumber of transcripts per cluster:')
print(f'cluster 1 = {fst_index_clust[0]}')

for i in range(1, length):
    print(f'cluster {i+1} = {fst_index_clust[i] - fst_index_clust[i-1]}')
print(f'cluster {i+2} = {len(df) - fst_index_clust[i]}')
```

## C) Heatmap of the clusters containing ATG7 isoforms

```python
%%time
#Open the GTEx info
path = input_file_directory + "/xenabrowser_brut_data/GTEx_info.tsv"
df_gtex_info = pandas.read_csv(path, sep = "\t", index_col='sample')

df_gtex_info = df_gtex_info[['_primary_site']]
list_tissues = set(df_gtex_info['_primary_site'])

print(df_gtex_info.shape)
df_gtex_info.head()
```

```python
# #reinstall the cluster column
# df['xcluster'] = cluster_uniq
cluster_atg7_1 = df.loc[df.index == ATG7_1]['xcluster'][0]
cluster_atg7_2 = df.loc[df.index == ATG7_2]['xcluster'][0]

#create df with the entire cluster where iso1 and iso2 are
df_2_clusters =  df[ (df['xcluster'] == cluster_atg7_1) | (df['xcluster'] == cluster_atg7_2) ]
print( set(df_2_clusters['xcluster']) )

df_2_clusters.head(5)
```

```python
#pop the cluster column
cluster_uniq = df_2_clusters.pop("xcluster")

#combine the two df
print(df_2_clusters.T.shape)
df_2_clusters = pandas.merge(df_gtex_info, df_2_clusters.T, left_index=True, right_index=True)
print(df_2_clusters.shape)

df_2_clusters = df_2_clusters.sort_values('_primary_site')
df_2_clusters.head(5)
```

```python
#Drop the two first rows because <not provided>
df_2_clusters.drop(df_2_clusters.index[0:2], inplace=True)

#save the list of tissues
list_tissues = set(df_2_clusters['_primary_site'])
print(len(list_tissues))

#Transpose back to have transcripts in index
df_2_clusters = df_2_clusters.T
df_2_clusters.columns = df_2_clusters.iloc[0]
df_2_clusters.drop(df_2_clusters.index[0], inplace=True)

df_2_clusters = df_2_clusters.astype(float)
print(df_2_clusters.shape)
df_2_clusters.head()
```

```python
# color for tissues. from hot to cold colors for A to Z 
seaborn.hls_palette(31)
```

```python
%%time
# no colors in columns
#use one color per isoform 
lut = dict(zip(cluster_uniq.unique(), seaborn.hls_palette(2)))
row_colors_clust = cluster_uniq.map(lut)

#plot
htm = seaborn.clustermap(df_2_clusters, cmap='bwr', col_cluster=True, row_cluster=False,
                   center=0, 
                   vmin=-6, vmax=6,
                   row_colors= row_colors_clust,
                   yticklabels= '', xticklabels = '',
                   cbar_kws={'label':'expression [log2(tpm+1)]'})
#save
title_file = 'heatmap_cluster_1_2_no_color_tissues'
plt.savefig(path2+title_file+'_article.tif')

plt.show()
plt.close()
```

```python
%%time
#use one color per isoform 
lut = dict(zip(cluster_uniq.unique(), seaborn.hls_palette(2)))
row_colors_clust = cluster_uniq.map(lut)

lut2 = dict(zip(list_tissues, seaborn.hls_palette(31)))
col_colors_tissues = df_2_clusters.columns.map(lut2)

#plot
htm = seaborn.clustermap(df_2_clusters, cmap='bwr', col_cluster=True, row_cluster=False,
                   center=0, 
                   vmin=-6, vmax=6,
                   row_colors= row_colors_clust,
                   col_colors= col_colors_tissues,
                   yticklabels= '', xticklabels = '',
                   cbar_kws={'label':'expression [log2(tpm+1)]'})
#save
title_file = 'heatmap_cluster_1_2'
plt.savefig(path2+title_file+'_article.tif')

plt.show()
plt.close()
```

```python
list_cluters_tissues = htm.dendrogram_col.reordered_ind
print(len(list_cluters_tissues), '\n')

data_tissues = df_2_clusters.columns

for number in list_cluters_tissues:
    tissue = data_tissues[number]
    print(tissue)
```

```python
#same the list of tissues in the cluster order
list_cluters_tissues = htm.dendrogram_col.reordered_ind
print(len(list_cluters_tissues), '\n')
col_tissue_clusters = []

#select the tissues
data_tissues = df_2_clusters.columns

for number in list_cluters_tissues:
    tissue = data_tissues[number]
    col_tissue_clusters.append(tissue)
    
lenght = len(list_cluters_tissues)

#select until the end of purple tissues (testis)
only_start = int(lenght*0.0755)
start_tissue_clusters = list_cluters_tissues[0:only_start]
print(len(start_tissue_clusters))

df_2_clusters_330firsts = df_2_clusters.iloc[:, start_tissue_clusters]
df_2_clusters_330firsts
```

```python
%%time
#use one color per isoform 
lut = dict(zip(cluster_uniq.unique(), seaborn.hls_palette(2)))
row_colors_clust = cluster_uniq.map(lut)

lut2 = dict(zip(list_tissues, seaborn.hls_palette(31)))
col_colors_tissues = df_2_clusters_330firsts.columns.map(lut2)

#plot
htm = seaborn.clustermap(df_2_clusters_330firsts, cmap='bwr', col_cluster=True, row_cluster=False,
                   center=0, 
                   vmin=-6, vmax=6,
                   row_colors= row_colors_clust,
                   col_colors= col_colors_tissues,
                   yticklabels= '', xticklabels = '',
                   cbar_kws={'label':'expression [log2(tpm+1)]'})
#save
title_file = 'heatmap_cluster_1_2_first_tissues'
plt.savefig(path2+title_file+'_article.tif')

plt.show()
plt.close()
```
