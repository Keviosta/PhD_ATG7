## all interactions

```python
df_inter_mefs_ori = pandas.read_csv(path+ 'mefs_all_interactions_mutants_cyto.csv', 
                                    sep = ";", decimal=',')
df_inter_mefs_ori.head()
```

## To group them, D522A/E (not *,  it means del)

```python
df_inter_mefs_ori['IsSTRING'] = "False"

df_inter_mefs_ori['Prot1'] = df_inter_mefs_ori['Bait']
df_inter_mefs_ori['Prot2'] = df_inter_mefs_ori['Gene']

df_inter_mefs = df_inter_mefs_ori[['Prot1' , 'Prot2', 'FoldChange', 'IsSTRING']]
df_inter_mefs.head()
```

```python
df_inter_mefs_gr = df_inter_mefs.copy()

df_inter_mefs_gr['Prot1'].replace('ATG7_1d522a', 'ATG7(1)-D522*', inplace=True)
df_inter_mefs_gr['Prot1'].replace('ATG7_1d522e', 'ATG7(1)-D522*', inplace=True)

df_inter_mefs_gr['Prot2'].replace('ATG7_2d522a', 'ATG7(2)-D522*', inplace=True)
df_inter_mefs_gr['Prot2'].replace('ATG7_2d522e', 'ATG7(2)-D522*', inplace=True)
```
