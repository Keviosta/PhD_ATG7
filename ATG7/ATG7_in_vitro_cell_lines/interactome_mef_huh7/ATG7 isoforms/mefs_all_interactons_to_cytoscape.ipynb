# MEFs

## all interactions

```python
df_inter_mefs_ori = pandas.read_csv(path+ 'mefs_all_interaction_both_iso_cyto.csv', sep = ",")
df_inter_mefs_ori.head()
```

```python
df_inter_mefs = pandas.DataFrame()

df_inter_mefs['Gene'] = df_inter_mefs_ori[['Gene']]

col_bait = df_inter_mefs_ori.iloc[:,17]
df_inter_mefs['Bait'] = col_bait

df_inter_mefs.rename(columns={df_inter_mefs.columns[0]: 'Prot1', 
                              df_inter_mefs.columns[1]: 'Prot2'},inplace=True)

df_inter_mefs['FoldChange'] = df_inter_mefs_ori[['FoldChange']]
df_inter_mefs['IsSTRING'] = "False"
df_inter_mefs.head()
```

## STRING data

```python
df_off = pandas.read_csv(path+ 'mefs_string_from_all_int_mouse.tsv', sep = "\t")
df_off.rename(columns={df_off.columns[0]: 'Prot1', df_off.columns[1]: 'Prot2'},inplace=True)
df_off = df_off[['Prot1', 'Prot2']]
df_off['IsSTRING'] = "True"
df_off.head()
```

```python
df_final = pandas.concat([df_inter_mefs, df_off], axis=0)
df_final
```

```python
df_final.to_csv(path+'df_cytoscape_mefs.csv',sep = ",")
```
