# Huh7

## all interactions

```python
df_inter_huh7_ori = pandas.read_csv(path+ 'huh7_all_interactions_both_iso_cyto.csv', sep = ",")
df_inter_huh7_ori.head()
```

```python
df_inter_huh7_ori['Prot1'] = df_inter_huh7_ori['name'].str.split(' ').str[3]
df_inter_huh7_ori['Prot2'] = df_inter_huh7_ori['name'].str.split(' ').str[0]
df_inter_huh7_ori['IsSTRING'] = "False"

df_inter_huh7 = df_inter_huh7_ori[['Prot1' , 'Prot2', 'FoldChange', 'IsSTRING']]
df_inter_huh7.head()
```

## STRING data

```python
df_off = pandas.read_csv(path+ 'huh7_string_from_all_int_human.tsv', sep = "\t")
df_off.rename(columns={df_off.columns[0]: 'Prot1', df_off.columns[1]: 'Prot2'},inplace=True)
df_off = df_off[['Prot1', 'Prot2']]
df_off['IsSTRING'] = "True"
df_off.head()
```

```python
df_final = pandas.concat([df_inter_huh7, df_off], axis=0)
df_final
```

```python
df_final.to_csv(path2+'df_cytoscape_huh7.csv',sep = ",")
```
