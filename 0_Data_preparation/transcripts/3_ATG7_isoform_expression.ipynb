# II] Data download

```python
%%time
# DL data all transcripts expression
df_ori = pandas.read_csv(path+'all_transcripts/gtex_Kallisto_tpm.tsv', sep = '\t', index_col=0)
print(df_ori.shape)
df_ori.head()
```

```python
# DL list selected samples
filtered_samples_list = numpy.loadtxt(path2+'filtered_samples_list.tsv', delimiter="\t", dtype=str)
len(filtered_samples_list)
```

```python
# DL phenotype GTEX info from: https://xenabrowser.net/datapages/?dataset=GTEX_phenotype&host=https%3A%2F%2Ftoil.xenahubs.net&removeHub=https%3A%2F%2Fxena.treehouse.gi.ucsc.edu%3A443
GTEX_info = pandas.read_csv(path+'GTEX_info.tsv', sep = '\t', index_col=0)
GTEX_info = GTEX_info[['_primary_site']]

GTEX_info.head()
```

# III] Transform data

## A. Subset

```python
ATG7_1 = 'ENST00000354449.7'
ATG7_2 = 'ENST00000354956.9'
ATG7_3 = 'ENST00000446450.6'

df = df_ori.loc[[ATG7_1,ATG7_2,ATG7_3]]
print(df.shape)

df
```

```python
# Subset, only samples we selected
df = df[filtered_samples_list]
print(df.shape)

# add tissue information
df = df.T
df = pandas.concat([GTEX_info, df], axis=1)

# rename the transcripts
df.rename(columns={df.columns[0]: 'Primary_Site', 
                        df.columns[1]: 'ATG7_1',
                        df.columns[2]: 'ATG7_2',
                        df.columns[3]: 'ATG7_3'},inplace=True)


df = df.dropna(subset=['ATG7_1'])

df
```

## B. To log2(tpm+1)

```python
%%time
# transform data from log2(tpm + 0.001) to tpm
print(f'min in log2(tpm+0.001): {df.iloc[:,1:].min().min()}')
#to tpm
df.iloc[:,1:] = ((2**df.iloc[:,1:]) - 0.001)
print(f'min in tpm: {df.iloc[:,1:].min().min()}')

# if value is negative, it's zero
df.iloc[:,1:] = (df.iloc[:,1:]).where(df.iloc[:,1:] > 0, 0)
print(f'new min in tpm: {df.iloc[:,1:].min().min()}')

# save tpm in a separate df
df_tpm = df.iloc[:,1:]
```

```python
%%time
# transform data from tpm to log2(tpm + 1)
#to tpm +1
df.iloc[:,1:] = df.iloc[:,1:] + 1
print(f'min in tpm+1: {df.iloc[:,1:].min().min()}')

#to log2(tpm+1)
df.iloc[:,1:] = numpy.log2(df.iloc[:,1:]) 
print(f'min in log2(tpm+1): {df.iloc[:,1:].min().min()}')

df.head(2)
```

## C. ATG7 isoform preparation

```python
#Prepare the columns with TPM, percentage
# Column rename
df_tpm = df_tpm.rename(columns = {'ATG7_1' : 'ATG7_1tpm',
                                  'ATG7_2' : 'ATG7_2tpm',
                                  'ATG7_3' : 'ATG7_3tpm'})
 
# Calcul to see percentage of expression of the 3 isoforms of ATG7 
df_tpm = df_tpm.assign(ATG7_prot_tot= df_tpm['ATG7_1tpm'] + df_tpm['ATG7_2tpm'] + df_tpm['ATG7_3tpm'])

df_tpm = df_tpm.assign(ATG7_1perc= (df_tpm['ATG7_1tpm'] *100)/ df_tpm['ATG7_prot_tot'],
                       ATG7_2perc= (df_tpm['ATG7_2tpm'] *100)/ df_tpm['ATG7_prot_tot'],
                       ATG7_3perc= (df_tpm['ATG7_3tpm'] *100)/ df_tpm['ATG7_prot_tot'])                              

df_tpm.head(5)
```

```python
df = pandas.concat([df, df_tpm], axis=1)
df.head(5)
```

```python
# What tissues are in the data? 
df['Primary_Site'].value_counts()
```

```python
#Remove tissues with lower than 20 samples. 
element_to_delete = []

#select all the elements in primary sites, and note which one are <20
working_elements = df["Primary_Site"].to_list()
for element in working_elements:
    if working_elements.count(element) < 20:
        element_to_delete.append(element)

#Convert to Set() to have a clear list because set cannot contain duplicate. 
element_to_delete = set(element_to_delete)
print('elements with less than 20 samples are:', element_to_delete)

#Remove the samples where the tissue is represented less than 20 times
print('\nshape with all the elements', df.shape)
for element in element_to_delete:
    df = df.loc[df["Primary_Site"] != element]
print('shape after removing the elements', df.shape)

df.head(5)
```
